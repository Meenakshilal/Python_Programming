```python
def fun3(n):
    '''it will return list of all values of dictionary.
    This will work for 2 level nesting as well'''
    lst=[]
    fn3b(n,lst)
    return lst
def fn3a(x):
    '''it will return the values of dictionary'''
    return x.values()
def fn3b(z,lst):
    ''' it will unwrap the dictionary and return list of values '''
    if type(z)== dict:
        a= fn3a(z)
        for i in a:
            if type(i)==dict:
                b=fn3b(i,lst)
            else:
                lst.append(i)
```

```python
fun3({"a":4,"b":{"c":6,"v":8,"n":{"c":78,"t":56}}})
```

```python
def fun4(x):
    '''it will give list of numbers'''
    return list(range(x))
def fun4a(a):
    '''it will give the square of numbers and will append it in a list'''
    if type(a)== list:
        l1= []
        for i in a:
            l1.append(i**2)
    return l1
```

```python
def con(x):
    '''it will concatinate multiple lists together'''
    l=[]
    for i in x:
        if type(i)== list:
            for j in i:
                l.append(j)
            
    return l 
```

```python
z= ([1,2,3,5,6],(4,5,6,9),[78,5,6,96],4586,[45,56,12,3],'a','b')
```

```python
con(z)
```

```python
def fun5(x):
    '''It will return the index of elements in list 
    even of repeative element'''
    for i,j in enumerate(x):
            yield (i,j)
```

```python
v=[4,5,2,6,4,5,'ab','ba',7,7]
for k in fun5(v):
    print (k)
```

```python
def dt():
    '''it will return the current date and time'''
    from datetime import datetime
    now = datetime.now()
    d = now.strftime("%d/%m/%Y %H:%M:%S")
    return d
```

```python
def img(x):
    '''It will return the image input'''
    import matplotlib.image as mpic
    import matplotlib.pyplot as plt

    # Read Images
    imag = mpic.imread(x)

    # Output Images
    return plt.imshow(imag)
```
